## Empirical bootstrap is non-parametric technique i.e used when population distribution and  parameters i.e mean,variance is unknown

```
import numpy as np 
from sklearn.utils import resample
import matplotlib.pyplot as plt
```

```
def bootstrap_emperical_CI(samples,CT = np.median,CI = 0.95,n_iterations=1000,n_size = 100):

    """ 
    This function will calculate confidence interval for a statitic measure i.e mean,median etc Using a non-parametric 
    approach i.e bootstrapping.
    Parameters :
    samples : population sample
    CT : central tendency - numpy fn
    n_iterations : no of bootstraped samples we will generate
    n_size : size of bootstraped samples **** Increase sample size to get more accurate intervals

    returns :
    a tuple - (lower CI,upper CI)

    """
    ct_measure = list()

    for i in range(n_iterations):
        # prepare a bootstraped sample
        s = resample(samples,n_samples = n_size)
        m = CT(s)
        ct_measure.append(m)

    # plot medians 
    plt.hist(ct_measure)
    plt.style.use('ggplot')
    plt.title('plot of  central tendency of bootstraped samples')
    plt.show()
    # confidence interval
    alpha = 0.95 # probablity of estimation

    p = ((1.0-alpha)/2.0)*100

    lower = np.percentile(ct_measure,p)

    p = (alpha+((1.0-alpha)/2.0))*100 
    upper = np.percentile(ct_measure,p)

    return lower,upper

```
